```python
import numpy as np
import matplotlib.pyplot as plt
from linear_boosting import Linear_Model,Linear_Boost
```

```python
samples = 100

X_0 = np.linspace(0,100,samples) + np.random.randn(samples)*10
X_1 = np.arange(0,samples) + np.random.randn(samples)*10
X = np.vstack([X_0,X_1]).T
Y = np.arange(0,samples) + np.random.randn(samples)*10

plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
plt.scatter(X[:,0],Y)
plt.show()
```

```python
plt.scatter(X[:,1],Y)
plt.show()
```

```python
model = Linear_Boost(X,Y)
model.fit()
```

```python
print('r_squared single linear model: ',model.r_squared[0])
```

```python
print('r_squared boosted linear model ensemble: ',model.r_squared[model.best_model])
```
